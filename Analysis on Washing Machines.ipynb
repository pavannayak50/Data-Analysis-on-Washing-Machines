  ## from  soup.find_all    seperated the  Brands ,  Capacity   &  loads  by usig    split()

```python
brands = []
for i in soup.find_all("div",class_="_4rR01T"):
    a=i.text
    brands.append(a.split(" ")[0])
brands
```

```python
prices = []
for i in soup.find_all("div",class_="_30jeq3 _1_WHN1"):
    a=i.text
    prices.append(a.split(" ")[0])
prices
```

```python
data = []
for i in soup.find_all("li",class_="rgWa7D"):
    a=i.text
    data.append(a)
data
```

```python
import re
loads= []
for i in soup.find_all("li",class_="rgWa7D"):
    a = i.text
    b = re.findall(r"\w+\sAutomatic\sTop\sLoad",a)
    if b:
        loads.append(b)
loads
```

```python
import re
rpm= []
for i in soup.find_all("li",class_="rgWa7D"):
    a = i.text
    b = re.findall(r"\d\w+",a)
    if b:
        rpm.append(b)
rpm
```

```python
import re
rating= []
for i in soup.find_all("li",class_="rgWa7D"):
    a = i.text
    b = re.findall(r"\d\sStar\sRating+",a)
    if b:
        rating.append(b)
rating
```

```python
import re
warranty= []
for i in soup.find_all("li",class_="rgWa7D"):
    a = i.text
    b = re.findall(r"^\d\sYears",a)
    if b:
        warranty.append(b)
warranty
```

```python
url1="""https://www.flipkart.com/washing-machines/pr?sid=j9e%2Cabm%2C8qx&otracker=categorytree&page="""
Brands=[]
Prices=[]
Capacity=[]
RPM=[]
LOADS=[]
Rating=[]
Warranty=[]

for i in range(2,35):
    urls = url1+str(i)
    
    print(urls)
    page = requests.get(urls)
    soup = BeautifulSoup(page.text)
    
# prices = []
    for i in soup.find_all("div",class_="_30jeq3 _1_WHN1"):
        a=i.text
        prices.append(a.split(" ")[0])
# prices 

    for i in soup.find_all("div",class_="_4rR01T"):
        a=i.text
        brands.append(a.split(" ")[0])
    
#     for i in soup.find_all("div",class_="gUuXy-"):
#         a=i.text
#         stars.append(a.split(",")[0])    
        
    for i in soup.find_all("li",class_="rgWa7D"):
        a = i.text
        b = re.findall(r"\w+\sAutomatic\sTop\sLoad",a)
        if b:
            loads.append(b)
# rpm= []
    for i in soup.find_all("li",class_="rgWa7D"):
        a = i.text
        b = re.findall(r"\d\w+",a)
        if b:
            rpm.append(b)
        
# rating= []
    for i in soup.find_all("li",class_="rgWa7D"):
        a = i.text
        b = re.findall(r"\d\sStar\sRating+",a)
        if b:
            rating.append(b)
        
# capacity= []
    for i in soup.find_all("li",class_="rgWa7D"):
        a = i.text
        b = re.findall(r"\s\d.\d\skg|\s\d\skg",a)
        if b:
            capacity.append(b)
        
    for i in soup.find_all("li",class_="rgWa7D"):
        a = i.text
        b = re.findall(r"^\d\sYears",a)
        if b:
            warranty.append(b)
```
